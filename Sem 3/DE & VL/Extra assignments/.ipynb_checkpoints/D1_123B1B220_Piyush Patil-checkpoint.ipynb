```python
# Importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset (ensure the path is correct without extra quotes)
data = pd.read_csv('student_Grades.csv')

# Univariate Analysis: Plot histograms for each numerical feature
# List of numerical columns
numerical_columns = ['absence_days', 'weekly_self_study_hours', 'math_score', 
                     'history_score', 'physics_score', 'chemistry_score', 
                     'biology_score', 'english_score', 'geography_score']

# Plot histograms for each numerical column
plt.figure(figsize=(15, 10))
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(3, 3, i)
    sns.histplot(data[column], bins=10, kde=True)
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset (ensure the path is correct)
data = pd.read_csv('student_Grades.csv')

# List of numerical columns
numerical_columns = ['absence_days', 'weekly_self_study_hours', 'math_score', 
                     'history_score', 'physics_score', 'chemistry_score', 
                     'biology_score', 'english_score', 'geography_score']

# Box Plots for Each Numerical Feature
plt.figure(figsize=(15, 10))
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(3, 3, i)  # Using a 3x3 grid for 9 numerical columns
    sns.boxplot(y=data[column])
    plt.title(f'Box Plot of {column}')
    plt.xlabel(column)
plt.tight_layout()
plt.show()
```

```python
# Importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the student_score dataset
data = pd.read_csv('student_Grades.csv')

# Example: Scatter Plot: Weekly Self Study Hours vs Math Score
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='weekly_self_study_hours', y='math_score', hue='gender', style='gender')
plt.title('Weekly Self Study Hours vs Math Score')
plt.xlabel('Weekly Self Study Hours')
plt.ylabel('Math Score')
plt.show()
```

```python
# Importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the student_score dataset (Ensure the path is correct)
data = pd.read_csv('student_Grades.csv')

# Correlation Heatmap: Showing correlation between numeric variables
plt.figure(figsize=(10, 6))

# Select only numeric columns for correlation
numeric_columns = data.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = data[numeric_columns].corr()

# Create the heatmap
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Heatmap of Student Score Dataset')
plt.show()
```
