```python
# Copying the dataframe to avoid chained assignment issue
newdt = dataset.copy()

# Clean up column names by removing leading/trailing spaces
newdt.columns = newdt.columns.str.strip()

# Filling missing values in 'age' column with the mean age
newdt['age'] = newdt['age'].fillna(newdt['age'].mean())

# Check if 'unknown' is already a category in the 'deck' column
if 'unknown' not in newdt['deck'].cat.categories:
    # Add 'unknown' as a new category in 'deck'
    newdt['deck'] = newdt['deck'].cat.add_categories('unknown')

# Fill missing values in 'deck' with 'unknown'
newdt['deck'] = newdt['deck'].fillna('unknown')

# Fill missing values in 'embark_town' with 'sanfransisco'
newdt['embark_town'] = newdt['embark_town'].fillna('sanfransisco')

# Fill missing values in 'embarked' with 'S'
newdt['embarked'] = newdt['embarked'].fillna('S')

# Fill missing values in other columns
newdt['pclass'] = newdt['pclass'].fillna(1)
newdt['sex'] = newdt['sex'].fillna('male')
newdt['sibsp'] = newdt['sibsp'].fillna(0)
newdt['parch'] = newdt['parch'].fillna(0)
newdt['fare'] = newdt['fare'].fillna(newdt['fare'].mean())
newdt['class'] = newdt['class'].fillna('first')
newdt['who'] = newdt['who'].fillna('woman')
newdt['adult_male'] = newdt['adult_male'].fillna('false')
newdt['alive'] = newdt['alive'].fillna('no')
newdt['alone'] = newdt['alone'].fillna('False')

newdt
```
