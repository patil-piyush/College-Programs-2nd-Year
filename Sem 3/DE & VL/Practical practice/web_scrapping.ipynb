```python
import requests as rq
from bs4 import BeautifulSoup as bs

web = rq.get("https://en.wikipedia.org/wiki/C_(programming_language)")
print(web)

websoup = bs(web.text, 'html.parser') 
websoup.prettify()
```

```python
websoup.title

for heading in websoup.find_all('h1'):
    print(heading.text)

for heading in websoup.find_all('h1'):
    print(heading)
```
