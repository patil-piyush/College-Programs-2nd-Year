```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

data = pd.read_csv('tested.csv')
df = pd.DataFrame(data)
print(data.head())
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Dropping rows with missing 'Age' values for univariate analysis
age_data = data['Age'].dropna()

# 1. Statistical Summary
age_summary = age_data.describe()

# 2. Plotting Distribution

# Histogram with density plot
sns.histplot(age_data, bins=30, kde=True, color='skyblue', stat='density')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Density')

# Display statistical summary and plot
plt.show()

age_summary
```
