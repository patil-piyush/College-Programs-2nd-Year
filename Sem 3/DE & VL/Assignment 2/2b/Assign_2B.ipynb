```python
import requests as rq
from bs4 import BeautifulSoup as bs

# Get Request
web = rq.get('https://www.wikipedia.org/')
print(web)

# Parse the content with BeautifulSoup
soup = bs(web.text, 'html.parser')

# Find all tables on the page
content = soup.find_all('table')

# Check if there are any tables
if len(content) > 1:
    table_all = content[1]  # Access the second table (index 1)
    print(table_all)
else:
    print("No tables found on the page.")
```

```python
title2 = soup.find_all('li')
for title in title2:
    print(title.text)
```
